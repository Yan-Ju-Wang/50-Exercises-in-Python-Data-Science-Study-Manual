## 091. 建立 $x$ 與 $cos(x)$

定義函數 `create_x_cosx_ndarray()` 能夠回傳長度 100、介於 $0.5\pi$ 與 $2.5\pi$ 的 $x$ 與其對應的 $cos(x)$

- 使用 `np.pi` 常數。
- 使用 `np.linspace()` 函數。
- 使用 `np.cos()` 函數。
- 將預期輸出寫在 `return` 之後。

```python
def create_x_cosx_ndarray() -> tuple:
    """
    >>> x, cosx = create_x_cosx_ndarray()
    >>> type(x)
    numpy.ndarray
    >>> type(cosx)
    numpy.ndarray
    >>> x.size
    100
    >>> cosx.size
    100
    """
    ### BEGIN SOLUTION
    x_ndarray = np.linspace(np.pi * 0.5, np.pi * 2.5, 100)
    cos_ndarray = np.cos(x_ndarray)
    return x_ndarray, cos_ndarray
    ### END SOLUTION
```

```python
x, cosx = create_x_cosx_ndarray()
type(x)
type(cosx)
x.size
cosx.size
```

## 096. 加減微小值調整 $x$ 與 $tan(x)$

定義函數 `create_adjusted_x_tanx_ndarray()` 能夠回傳長度 10000、介於 $0.5\pi + epsilon$ 與 $1.5\pi - epsilon$ 的 $x$ 與其對應的 $tan(x)$，其中 $epsilon=10^{-6}$

- 使用 `np.pi` 常數。
- 使用 `np.linspace()` 函數。
- 使用 `np.tan()` 函數。
- 將預期輸出寫在 `return` 之後。

```python
def create_adjusted_x_tanx_ndarray() -> tuple:
    """
    >>> adjusted_x, adjusted_tanx = create_adjusted_x_tanx_ndarray()
    >>> type(adjusted_x)
    numpy.ndarray
    >>> type(adjusted_tanx)
    numpy.ndarray
    >>> adjusted_x.size
    10000
    >>> adjusted_tanx.size
    10000
    >>> adjusted_x[0] > 0.5 * np.pi
    True
    >>> adjusted_x[-1] < 1.5 * np.pi
    True
    """
    ### BEGIN SOLUTION
    adjusted_x_ndarray = np.linspace((0.5 * np.pi + 10**-6), (1.5 * np.pi - 10**-6), 10000)
    adjusted_tanx_ndarray = np.tan(adjusted_x_ndarray)
    return adjusted_x_ndarray, adjusted_tanx_ndarray
    ### END SOLUTION
```

```python
adjusted_x, adjusted_tanx = create_adjusted_x_tanx_ndarray()
type(adjusted_x)
type(adjusted_tanx)
adjusted_x.size
adjusted_tanx.size
adjusted_x[0] > 0.5 * np.pi
adjusted_x[-1] < 1.5 * np.pi
```

## 100. 調整 y 軸上下限

定義函數 `export_x_tanx_png_final()` 能夠輸出介於 $0.5\pi + epsilon$ 與 $1.5\pi - epsilon$ 的 $x$ 與其對應的 $tan(x)$，其中 $epsilon=10^{-6}$，並且加上圖表標題、調整 y 軸上下限為 $[-5, 5]$，圖檔名為 `xtanx_final.png`

![](../test-images/xtanx_final.png)

- 使用 `create_adjusted_x_tanx_ndarray()` 函數
- 使用 `plt.subplots()` 函數。
- 使用 `AxesSubplot.plot()`
- 使用 `AxesSubplot.set_title()`
- 使用 `AxesSubplot.set_ylim()`
- 使用 `fig.savefig()`

```python
def export_x_tanx_png_final() -> None:
    """
    >>> export_x_tanx_png_final()
    >>> im = Image.open("xtanx_final.png")
    >>> im.format
    'PNG'
    """
    ### BEGIN SOLUTION
    adjusted_x_ndarray, adjusted_tanx_ndarray = create_adjusted_x_tanx_ndarray() 
    fig, ax = plt.subplots()
    ax.plot(adjusted_x_ndarray, adjusted_tanx_ndarray)
    ax.set_title('tan(x)')
    ax.set_ylim(-5, 5)
    fig.savefig('xtanx_final.png')
    ### END SOLUTION
```
